```python
#creating code to Plot sin Curve.

#importing numpy for data Manipulation

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi,30)
y = np.sin(x)

plt.plot(x,y,
        color ='red',  # line color
        linestyle='--',
        linewidth=3,
        marker='o',
        markersize=5,
        markeredgecolor='black',
        markerfacecolor='green',
        markeredgewidth='1',  #edge width
        alpha=1,           # Transparency
        label='Data - line')  #label for Legend

plt.xlabel("Time",color='green')
plt.ylabel("Amplitude",color='Blue')
plt.title("Simple Harmonic Motion curve", color = 'red')
plt.grid()
plt.legend()
plt.show()
```
