```python
# creating more than one Axes plot

import matplotlib.pyplot as plt

import numpy as np

# create figure with two axes

fig, (ax1 , ax2,ax3,ax4) = plt.subplots(nrows = 4, ncols = 1, figsize =(8,8))

# plotting graph on first axes

x1 = np.linspace(0, 10,100 )
y1 = np.sin(x1)
ax1.plot(x1,y1)
ax1.set_title("Sine Curve")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.grid

x2 = np.linspace(0, 40,100 )
y2 = np.tan(x2)
ax2.plot(x2,y2)
ax2.set_title("Tan Curve")
plt.xlabel("Time",color='green')
plt.ylabel("Amplitude")
plt.grid

x3 = np.linspace(0, 10,100 )
y3 = np.sin(x3)
ax3.plot(x3,y3)
ax3.set_title("Sec Curve")
plt.xlabel("Time",color='green')
plt.ylabel("Amplitude")
plt.grid

x4 = np.linspace(0, 10,100 )
y4 = np.cos(x4)
ax4.plot(x4,y4)
ax2.set_title("Cos Curve")
plt.xlabel("Time",color='green')
plt.ylabel("Amplitude")

plt.grid()
plt.show()
```
