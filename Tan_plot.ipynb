```python
# Code for Tan function

import numpy as np

import matplotlib.pyplot as plt

# defininf x asis evenly spaced values

x = np.linspace(0,4*np.pi,50)
y = np.tan(x)

plt.plot(x,y,
         color ='red',  # line color
        linestyle='--',
        linewidth=3,
        marker='o',
        markersize=5,
        markeredgecolor='black',
        markerfacecolor='green',
        markeredgewidth='1',  #edge width
        alpha=1,           # Transparency
        label='Data - line')  #label for Legend

plt.xlabel("Time ")
plt.ylabel("Amplitude")
plt.title("Tangent Graph",color= 'green')
plt.grid()
plt.legend()
plt.show()


```
