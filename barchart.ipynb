```python
#Simple Vertical bar chart

#A bar chart is a suitable choice for visualizing and comparing categorical data
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('/content/stocks.csv')
#print(data)

# Extract the 'date' and 'Apple' columns
dates = data['date']
#print(dates)
apple_prices = data['Apple']
#print(apple_prices)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(dates, apple_prices, color='skyblue')
plt.xlabel('Date')
plt.ylabel('Apple Stock Price')
plt.title('Apple Stock Prices in January 2022')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the chart
plt.show()
```
