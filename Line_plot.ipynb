```python
import matplotlib.pyplot as plt

# Define some data

x = [1,3,7,9]
y = [4,9,2,5]

#creating fig and Axes Object

fig, ax = plt.subplots()

#plotting data on axes

ax.plot(x,y,
        color ='red',  # line color
        linestyle='--',
        linewidth=2,
        marker='o',
        markersize=8,
        markeredgecolor='black',
        markerfacecolor='green',
        markeredgewidth='2',  #edge width
        alpha=0.5,           # Transparency
        label='Data - line')  #label for Legend

#defining X label and Y label

ax.set_xlabel("Time",color='blue')
ax.set_ylabel("Distance covered",color='green')

#setting Tittle
ax.set_title("Motion of an Object",color='magenta')

plt.grid()

plt.legend()
#code to Show Plot

plt.show()

```
